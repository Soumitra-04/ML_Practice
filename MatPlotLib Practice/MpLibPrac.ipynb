```python
income = [55,56,63,68,74,74,80,84]
income
```

```python
income_ticks = list(range(50,86,2))
income_ticks
```

```python
plt.title("Income of John (in USD)",fontsize = 15, fontname = "Times New Roman")
plt.xlabel("Year")
plt.ylabel("Yearly Income (in USD)")
plt.plot(years,income)
plt.yticks(income_ticks,[f"{x}k $" for x in income_ticks])
plt.show()
```
