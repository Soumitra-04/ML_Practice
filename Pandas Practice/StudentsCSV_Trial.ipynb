```python
import numpy as np
import pandas as pd
```

```python
my_df = pd.read_csv('students_data.csv')
my_df
```

```python
#counting Distinct values - Descending
my_df['passed'].value_counts()
```

```python
#counting Distinct values - Ascending
my_df['passed'].value_counts(ascending = True)
```

```python
#counting Distinct values - inclunding NaN
my_df['passed'].value_counts(dropna = False)
```

```python
#Getting Relative Frequency - Percentage
my_df['passed'].value_counts(normalize = True,dropna = False)
```

```python
#getting specific item count
my_df['passed'].value_counts()[1]
```

```python
#Counting Unique Values - Size
my_df.groupby('passed').size()
```

```python
#Counting Unique Values - Count
my_df.groupby('passed').count()
```

```python
#Get a count of all columns across all columns
my_df.select_dtypes(include='number').apply(lambda col: col.value_counts())
```

```python
my_df['attendance'].max()
```
